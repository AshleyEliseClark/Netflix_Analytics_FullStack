```python
import pandas as pd
import numpy as np
import os

data_path = r"C:\Users\AshleyPC\Netflix_Analytics_Project\1. data\raw"

content_df = pd.read_csv(os.path.join(data_path, 'content.csv'))
ratings_df = pd.read_csv(os.path.join(data_path, 'ratings.csv'))
users_df = pd.read_csv(os.path.join(data_path, 'users.csv'))
view_df = pd.read_csv(os.path.join(data_path, 'view_history.csv'))

print("ðŸ“„ Content Dataset:")
print(content_df.head())

print("ðŸ“Š Ratings Dataset:")
print(ratings_df.head())

print("ðŸ‘¤ Users Dataset:")
print(users_df.head())

print("ðŸŽ¬ View History Dataset:")
print(view_df.head())
```

```python
# Inspect structure and missing values
def check_df(df, name):
    print(f"\nðŸ§¹ Overview of {name} dataset:")
    print(df.info())
    print("\nMissing Values:")
    print(df.isnull().sum())
    print("-" * 50)

check_df(content_df, "Content")
check_df(ratings_df, "Ratings")
check_df(users_df, "Users")
check_df(view_df, "View History")
```

```python
# Convert string dates to datetime objects
ratings_df['rating_date'] = pd.to_datetime(ratings_df['rating_date'])
users_df['signup_date'] = pd.to_datetime(users_df['signup_date'])
view_df['watch_date'] = pd.to_datetime(view_df['watch_date'])
```
